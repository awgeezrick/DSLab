```python
import pandas as pd
import numpy as np
import xgboost as abg
from sklearn.model_selection import train_test_split, GridSearchCV
from scipy import stats
```

```python
xtrain = pd.read_csv("./draftlogs.csv")
n = xtrain.shape[0]
labels = xtrain.Y.values
labels = labels - 447137
x, x_test, y, y_test = train_test_split(xtrain,labels,test_size=0.2,train_size=0.8)
x_train, x_cv, y_train, y_cv = train_test_split(x,y,test_size = 0.25,train_size =0.75)

```

```python
xgbc_params = {
    'objective' : 'multi:softprob',
    'eval_metric': "merror",
    'num_class': 281,
    'n_jobs': -1,
    'learning_rate':0.1,
    'n_estimators':100,
    'max_depth':5,
    'min_child_weight':1,
    'gamma':0,
    'subsample':.8,
    'colsample_bytree':.8,
    'seed':69
}

train_matrix = abg.DMatrix(x.values, y)

xgb_clf_softprob = abg.train(xgbc_params, train_matrix)
```

```python
test_matrix = abg.DMatrix(x_test.values, y_test)
```

```python
# Note: this convention has been changed since xgboost-unity
# get prediction, this is in 1D array, need reshape to (ndata, nclass)
pred_prob = xgb_clf_softprob.predict(test_matrix).reshape(y_test.shape[0], 281)
pred_label = np.argmax(pred_prob, axis=1)
error_rate = np.sum(pred_label != y_test) / y_test.shape[0]
print('Test error using softprob = {}'.format(error_rate))
```

```python
xgbc_params = {
    'objective' : 'multi:softmax',
    'eval_metric': "merror",
    'num_class': 281,
    'n_jobs': -1,
    'learning_rate':0.1,
    'n_estimators':100,
    'max_depth':5,
    'min_child_weight':1,
    'gamma':0,
    'subsample':.8,
    'colsample_bytree':.8,
    'seed':69
}

xgb_clf_softmax = abg.train(xgbc_params, train_matrix)
```

```python
pred = xgb_clf_softmax.predict(test_matrix)
```

```python
error_rate = np.sum(pred != y_test) / y_test.shape[0]
print('Test error using softmax = {}'.format(error_rate))
```
